```python
# Import dependencies
import pandas as pd
import gmaps
import requests

# Import the API key.
from config import g_key

# Configure gmaps to use your Google API key.
gmaps.configure(api_key=g_key)
```

```python
# Load the CSV created in Weather_Database.ipynb into a DataFrame.
city_weather_data_df = pd.read_csv("weather_data/WeatherPy_database.csv", index_col = 0)
city_weather_data_df.head(10)
```

```python
# Filter dataframe for min and max temperature
# Ask customer for min and max temp preferences
min_temp = float(input("What is the minimum temperature you would like for your trip? "))
max_temp = float(input("What is the maximum temperature you would like for your trip? "))

# Ask customer if they want it to be raining
rain = str(input("Do you want it to be raining? (yes/no) "))

# Ask customer if they want it to be snowing
snow = str(input("Do you want it to be snowing? (yes/no) "))
```

```python
# Filter the dataset to find the cities that fit the temperature range criteria.
preferred_cities_df = city_weather_data_df.loc[(city_weather_data_df["Max Temp"] <= max_temp) & (city_weather_data_df["Max Temp"] >= min_temp)]
preferred_cities_df.head(10)
```

```python
# Filter the dataset to find cities with rain and snow preference.
if rain == 'no':
    preferred_cities_df = preferred_cities_df.loc[(preferred_cities_df["Rain (inches)"] == 0)]
elif rain == 'yes':
    preferred_cities_df = preferred_cities_df.loc[(preferred_cities_df["Rain (inches)"] != 0)]
elif snow == 'no':
    preferred_cities_df = preferred_cities_df.loc[(preferred_cities_df["Snow (inches)"] == 0)]
else:
    preferred_cities_df = preferred_cities_df.loc[(preferred_cities_df["Snow (inches)"] != 0)]
```

```python
# Add cities to a marker layer map with a pop-up marker for each city
# Create DataFrame called hotel_df to store hotel names along with city, country, max temp, and coordinates.
hotel_df = preferred_cities_df[["City", "Country", "Max Temp", "Current Description", "Lat", "Lng"]].copy()
hotel_df["Hotel Name"] = ""
hotel_df.head(10)
```

```python
# Set parameters to search for a hotel.
params = {
    "radius": 5000,
    "type": "lodging",
    "key": g_key
}
    
# Use base URL to search for hotels
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

# Make request and get the JSON data from the search.
hotels = requests.get(base_url, params=params).json()
```

```python
# Iterate through the DataFrame.
for index, row in hotel_df.iterrows():
    # Get the latitude and longitude.
    lat = row["Lat"]
    lng = row["Lng"]

    # Add the latitude and longitude to location key for the params dictionary.
    params["location"] = f"{lat},{lng}"

    # Use the search term: "lodging" and our latitude and longitude.
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    # Make request and get the JSON data from the search.
    hotels = requests.get(base_url, params=params).json()
    # Grab the first hotel from the results and store the name.
    try:
        hotel_df.loc[index, "Hotel Name"] = hotels["results"][0]["name"]
    except (IndexError):
        print("Hotel not found... skipping.")
```

```python
hotel_df.head(10)
```

```python
# Remove cities which do not have information on Hotels.
hotel_df.drop(hotel_df[(hotel_df['Hotel Name'] == "")].index, inplace = True)

hotel_df.head(10)
```

```python
hotel_df.index.name = None
hotel_df.head(10)
```

```python
# Create the output file (CSV).
output_data_file = "weather_data/WeatherPy_vacation.csv"
# Export the Hotel Data into a CSV.
hotel_df.to_csv(output_data_file, index_label="City_ID")
```

```python
# Convert Max Temp type from float64 to string.
hotel_df['Max Temp'] = hotel_df['Max Temp'].astype(str)
```

```python
hotel_df.head(10)
```

```python
# Creating a new column with Current Weather by merging Current Description and Max Temp into one.
hotel_df["Current Weather"] = hotel_df["Current Description"] +" and "+ hotel_df["Max Temp"] +" " u"\N{DEGREE SIGN}""F"

# Dropping the columns for Current Description and Max Temp as the information has been replaced with Current Weather
hotel_df = hotel_df.drop(['Current Description', 'Max Temp'], axis=1)

# Rearranging the columns in DataFrame
hotel_df = hotel_df[["City", "Country", "Lat", "Lng", "Current Weather", "Hotel Name"]]
```

```python
# Creating a pop up marker for the DataFrame for layering
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current Weather</dt><dd>{Current Weather}</dd>
</dl>
"""
```

```python
# Store the DataFrame Row.
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
```

```python
# Add marker layer with pop up on map.
locations = hotel_df[["Lat", "Lng"]]
current_description = hotel_df["Current Weather"]

fig = gmaps.figure(center=(20.0, 40.0), zoom_level=1.5)

marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)

#  Adding marker.
fig.add_layer(marker_layer)


# Call the figure to plot the data.
fig
```
